The transformers library was installed successfully.
This library helps us load and use the GPT-2 model.


```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import torch
```

The required Python modules were imported.
These will be used to load the GPT-2 model and generate text.


```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
```

The GPT-2 tokenizer and model were loaded successfully.
We are using a pre-trained model, so no training is required.


```python
niche = input("Enter YouTube niche: ")
```

The system now accepts a niche from the user.

```python
prompt = f"List of viral video titles about  {niche}:\n1.\n"
print(prompt)
```

A prompt was created using the user input.
This prompt tells GPT-2 what type of text to generate.


```python
input_ids = tokenizer.encode(prompt, return_tensors="pt")
```

The text prompt was converted into tokens.
GPT-2 works only with tokenized input.


```python
output = model.generate(
    input_ids,
    max_length=100,
    num_return_sequences=1,
    no_repeat_ngram_size=2,
    do_sample=True,
    top_k=50,
    top_p=0.95
)
```

GPT-2 generated text based on the given prompt.
The model uses sampling to create creative titles.


```python
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

The generated output was converted back into readable text.


The generated output is not perfectly structured as video titles.
Some lines are short or irrelevant, and some lines look like paragraph text instead of titles.

This happens because GPT-2 is a basic text generation model and not instruction-based.
It continues text based on patterns it has seen during training, rather than strictly following commands.

However, the model still produces topic-related content about Forex Trading.
By adjusting the prompt format and cleaning the output, useful video title ideas can be extracted.
